```python
import numpy as np
import scipy as sp
import scipy.sparse as sparse
import dynamical_networks as dynets
import matplotlib.pyplot as plt
import generate_adjacency_matrix as gen_adj
```

```python
from scipy.signal import sawtooth
from scipy.signal import square
from __future__ import division
```

```python
G_ON=1.
G_OFF=1e-2

def hard_switching_AS_G(G, voltage, threshold_voltage, G_ON, G_OFF):
    if voltage > threshold_voltage:
        return G_OFF
    elif voltage < -1. * threshold_voltage:
        return G_ON
    else:
        return G

def hard_switching_lindecay_AS_G(G, voltage, threshold_voltage, G_ON, G_OFF):
    if voltage > threshold_voltage:
        return G_OFF
    elif voltage < -1. * threshold_voltage:
        return G_ON
    else:
        return G - G_ON/4000.

def hard_switching_expdecay_AS_G(G, voltage, threshold_voltage, G_ON, G_OFF):
    if voltage > threshold_voltage:
        return G_OFF
    elif voltage < -1. * threshold_voltage:
        return G_ON
    else:
        return 0.99*G
    
def XtoG(X):
    return X + X.T
```

```python
num_nodes = 100
```

```python
steps_per_period  = 500
periods = 10
V_peak = 10
t = np.linspace(0, 1, periods * steps_per_period)

```

### First, A constant voltage

```python
Adj_cubic = gen_adj.cubic_2d((10, 10), undirected=False, xbias=.8, ybias=.8)
rows, cols = Adj_cubic.nonzero()
thresholds_vec = np.random.rand(len(rows))
threshold_voltages = sparse.csr_matrix((thresholds_vec, (rows, cols)))
M_init = Adj_cubic * G_OFF
v_initial = np.empty((num_nodes,)) * np.nan
v_initial[0] = 10
v_initial[-1] = 0

net_cubic = dynets.DynamicalNetwork(M_init.copy(), v_initial, XtoG, hard_switching_lindecay_AS_G,
                           threshold_voltages=threshold_voltages, G_ON=1, G_OFF=1e-2)
net_cubic.solve_voltages('cg')
```

```python
fig, ax = plt.subplots(nrows=10, ncols=5, figsize=(30, 60))
ax_counter=0
ax_step = 100//5
# Initialize the current array
input_current = []

#Now for each time point
for step in range(1000):
    #Solve for voltages
    net_cubic.solve_voltages('cg')
    #find the input current
    ext_current = net_cubic.external_current()
    input_current.append(ext_current[ext_current>0].sum())
    #Integrate the memristances for the next step
    net_cubic.update_state_variables()
    
    if step % ax_step == 0:
        net_cubic.display_grid(ax[ax_counter//5][ax_counter%5], (10, 10), 'conductance')
        ax_counter += 1
```

```python
fig, ax = plt.subplots(figsize=(16, 5))

ax.plot(range(1000), 10* np.ones((1000,)), 'r.')
ax.set_xlabel('time (s)')
ax.set_ylabel('Voltage (V)', color='r')
for t1 in ax.get_yticklabels():
    t1.set_color('r')
    
ax2 = ax.twinx()
ax2.plot(range(1000), input_current, 'b-')
ax2.set_ylabel('Current', color='b')
for t2 in ax2.get_yticklabels():
    t2.set_color('b')
```

### Pulses

```python
V_peak = 4.
V_high = V_peak / 2. * (square(t * 2 * np.pi * periods) + 1)
v_external = np.empty((num_nodes, periods*steps_per_period)) * np.nan
v_external[0, :] = V_high
v_external[-1, :] = 0.
```

```python
net_cubic = dynets.DynamicalNetwork(M_init.copy(), v_initial, XtoG, hard_switching_lindecay_AS_G,
                           threshold_voltages=threshold_voltages, G_ON=1, G_OFF=1e-2)
net_cubic.solve_voltages('cg')
```

```python

# Initialize the current array
input_current = []

#Now for each time point
for step in range(int(periods * steps_per_period)):
    #Update external voltages
    net_cubic.external_voltages = v_external[:,step]
    #Solve for voltages
    net_cubic.solve_voltages('cg')
    #find the input current
    ext_current = net_cubic.external_current()
    input_current.append(ext_current[ext_current>0].sum())
    #Integrate the memristances for the next step
    net_cubic.update_state_variables()
```

```python
fig, ax = plt.subplots(figsize=(15, 5))

ax.plot(range(int(periods * steps_per_period)), V_high, 'r.')
ax.set_xlabel('time (s)')
ax.set_ylabel('Voltage (V)', color='r')
for t1 in ax.get_yticklabels():
    t1.set_color('r')
    
ax2 = ax.twinx()
ax2.plot(range(int(periods * steps_per_period)), input_current, 'b-')
ax2.set_ylabel('Current', color='b')
for t2 in ax2.get_yticklabels():
    t2.set_color('b')
```

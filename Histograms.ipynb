Above we have the timecourses for 3 8x8 networks run at varying voltages.  When the voltages is too low to create a breakdown in the network it moves to a state of maximal conductance and above some threshold, the dynamics are jumps between between conductance states.

```python
def residence_times_cond(current, voltage, t, cutoff=0, tolerance=1e-5):
    """
    Returns a list of time intervals during which the conductance of the device was constant within a given tolerance.
    A cutoff may be introduced and should be at least the temporal spacing.
    """
    def conductance(index):
        if voltage[index] != 0:
            return current[index] / voltage[index]
        else:
            return np.nan
    
    res_times = []
    beginnings = []
    endings = []
    beginning = 0

    for idx, time in enumerate(t[:-1]):
        
        # a new conductance state is initialized if the conductance goes from nan to a new value
        if np.isnan(conductance(idx-1)) and not np.isnan(conductance(idx)):
            beginning = idx
        
        # We exit the state if the conductance changes to a value further away than the tolerance
        # or if it becomes nan
        if np.abs(conductance(idx+1) - conductance(beginning)) > tolerance or \
        (np.isnan(conductance(idx+1)) and not np.isnan(conductance(idx))):
            state_time = t[idx] - t[beginning]
            # only include it if the length of the state is longer than the cutoff.
            if state_time > cutoff:
                res_times.append(state_time)
                beginnings.append(beginning)
                endings.append(idx)
            beginning = idx+1
    
    return res_times, beginnings, endings
```

```python
res1, beg, end = residence_times_cond(current_1[2000:], v1[2000:-1], t1[2000:-1])
res2, beg, end = residence_times_cond(current_2[2000:], v2[2000:-1], t2[2000:-1])
res3, beg, end = residence_times_cond(current_3[2000:], v3[2000:-1], t3[2000:-1])
```

```python
hist1, bin1 = np.histogram(res1, bins=30)
hist2, bin2 = np.histogram(res2, bins=30)
hist3, bin3 = np.histogram(res3, bins=30)
fig, ax = plt.subplots(nrows=2, ncols=4, figsize=(16,8))
ax[0,0].plot(bin1[:-1], hist1)
ax[0,0].set_ylim(0, np.max(hist2))
ax[0,1].plot(bin2[:-1], hist2, 'g')
ax[0,1].set_ylim(0, np.max(hist2))
ax[0,2].plot(bin3[:-1], hist3, 'r')
ax[0,2].set_ylim(0, np.max(hist3))
ax[0,3].plot(bin1[:-1], hist1)
ax[0,3].plot(bin2[:-1], hist2, 'g')
ax[0,3].plot(bin3[:-1], hist3, 'r')

ax[1,0].scatter(bin1[:-1], hist1, c='b')
ax[1,1].scatter(bin2[:-1], hist2, c='g')
ax[1,2].scatter(bin3[:-1], hist3, c='r')
ax[1,0].set_yscale('log')
ax[1,0].set_xscale('log')
ax[1,0].set_ylim(bottom=1, top=hist1.max()*2)
ax[1,0].set_xlim(left=bin1.min()/2, right=bin1.max()*2 )
ax[1,1].set_yscale('log')
ax[1,1].set_xscale('log')
ax[1,1].set_ylim(bottom=1, top=hist2.max()*2)
ax[1,1].set_xlim(left=bin2.min()/2, right=bin2.max()*2 )
ax[1,2].set_yscale('log')
ax[1,2].set_xscale('log')
ax[1,2].set_ylim(bottom=1, top=hist3.max()*2)
ax[1,2].set_xlim(left=bin3.min()/2, right=bin3.max()*2 )
ax[1,3].scatter(bin1[:-1], hist1, c='b')
ax[1,3].scatter(bin2[:-1], hist2, c='g')
ax[1,3].scatter(bin3[:-1], hist3, c='r')
ax[1,3].set_yscale('log')
ax[1,3].set_xscale('log')
ax[1,3].set_ylim(bottom=1, top=np.amax(np.maximum(hist1, hist2, hist3))*2)
ax[1,3].set_xlim(left=np.amin(np.minimum(bin1, bin2, bin3))/2, right=np.amax(np.maximum(bin1, bin2, bin3))*2 )
plt.tight_layout()
```

```python
def residence_times_cond(current, voltage, t, cutoff=0, tolerance=1e-5):
    """
    Returns a list of time intervals during which the conductance of the device was constant within a given tolerance.
    A cutoff may be introduced and should be at least the temporal spacing.
    """
    def conductance(index):
        if voltage[index] != 0:
            return current[index] / voltage[index]
        else:
            return np.nan
    
    res_times = []
    beginnings = []
    endings = []
    beginning = 0

    for idx, time in enumerate(t[:-1]):
        
        # a new conductance state is initialized if the conductance goes from nan to a new value
        if np.isnan(conductance(idx-1)) and not np.isnan(conductance(idx)):
            beginning = idx
        
        # We exit the state if the conductance changes to a value further away than the tolerance
        # or if it becomes nan
        if np.abs(conductance(idx+1) - conductance(beginning)) > tolerance or \
        (np.isnan(conductance(idx+1)) and not np.isnan(conductance(idx))):
            state_time = t[idx] - t[beginning]
            # only include it if the length of the state is longer than the cutoff.
            if state_time > cutoff:
                res_times.append(state_time)
                beginnings.append(beginning)
                endings.append(idx)
            beginning = idx+1
    
    return res_times, beginnings, endings
```

```python
res1, beg, end = residence_times_cond(current_1[2000:], v1[2000:-1], t1[2000:-1])
res2, beg, end = residence_times_cond(current_2[10000:], v2[10000:-1], t2[10000:-1])
res3, beg, end = residence_times_cond(current_3[10000:], v3[10000:-1], t3[10000:-1])
```

```python
hist1, bin1 = np.histogram(res1, bins=50)
hist2, bin2 = np.histogram(res2, bins=50)
hist3, bin3 = np.histogram(res3, bins=50)
# pdf is the histogram divided by its sum
pdf1 = hist1 / np.sum(hist1)
pdf2 = hist2 / np.sum(hist2)
pdf3 = hist3 / np.sum(hist3)
# Cumulative distribution function is integral of the pdf
cdf1 = np.cumsum(pdf1)
cdf2 = np.cumsum(pdf2)
cdf3 = np.cumsum(pdf3)
# complementary cdf is 1 - cdf
ccdf1 = 1 - cdf1 
ccdf2 = 1 - cdf2
ccdf3 = 1 - cdf3

fig, ax = plt.subplots(nrows=2, ncols=4, figsize=(16,8))
ax[0,0].plot(bin1[:-1], pdf1, 'b-.')
ax[0,0].plot(bin1[:-1], ccdf1, 'b-')
ax[0,0].set_ylim(0, 1)
ax[0,1].plot(bin2[:-1], pdf2, 'g-.')
ax[0,1].plot(bin2[:-1], ccdf2, 'g-')
ax[0,1].set_ylim(0, 1)
ax[0,2].plot(bin3[:-1], pdf3, 'r-.')
ax[0,2].plot(bin3[:-1], ccdf3, 'r-')
ax[0,2].set_ylim(0, 1)
ax[0,3].plot(bin1[:-1], ccdf1)
ax[0,3].plot(bin2[:-1], ccdf2, 'g')
ax[0,3].plot(bin3[:-1], ccdf3, 'r')
ax[0,3].set_ylim(0, 0.3)

ax[1,0].scatter(bin1[:-1], ccdf1, c='b')
ax[1,0].scatter(bin1[:-1], pdf1, s=10, c='b', marker='^', alpha=0.5)
ax[1,1].scatter(bin2[:-1], ccdf2, c='g')
ax[1,1].scatter(bin2[:-1], pdf2, s=10, c='g', marker='^', alpha=0.5)
ax[1,2].scatter(bin3[:-1], ccdf3, c='r')
ax[1,2].scatter(bin3[:-1], pdf3, s=10, c='r', marker='^', alpha=0.5)
ax[1,0].set_yscale('log')
ax[1,0].set_xscale('log')
ax[1,0].set_ylim(bottom=1e-3, top=1)
ax[1,0].set_xlim(left=bin1.min()/2, right=bin1.max()*2 )
ax[1,1].set_yscale('log')
ax[1,1].set_xscale('log')
ax[1,1].set_ylim(bottom=1e-3, top=1)
ax[1,1].set_xlim(left=bin2.min()/2, right=bin2.max()*2 )
ax[1,2].set_yscale('log')
ax[1,2].set_xscale('log')
ax[1,2].set_ylim(bottom=1e-3, top=1)
ax[1,2].set_xlim(left=bin3.min()/2, right=bin3.max()*2 )
ax[1,3].scatter(bin1[:-1], ccdf1, c='b')
ax[1,3].scatter(bin2[:-1], ccdf2, c='g')
ax[1,3].scatter(bin3[:-1], ccdf3, c='r')
ax[1,3].set_yscale('log')
ax[1,3].set_xscale('log')
ax[1,3].set_ylim(bottom=1e-3, top=1)
ax[1,3].set_xlim(left=np.amin(np.minimum(bin1, bin2, bin3))/2, right=np.amax(np.maximum(bin1, bin2, bin3))*2 )
plt.tight_layout()
```

```python
from scipy.optimize import curve_fit

def exp_func(x, A, alpha):
    return A * np.exp(-1 * alpha * x)

def power_func(x, A, alpha):
    return A / (x ** alpha)

popt_exp2, pcov_exp = curve_fit(exp_func, bin2[:-1], ccdf2)
popt_pow2, pcov_pow = curve_fit(power_func, bin2[-31:-1], ccdf2[-30:])

popt_exp3, pcov_exp = curve_fit(exp_func, bin3[:-1], ccdf3)
popt_pow3, pcov_pow = curve_fit(power_func, bin3[-31:-1], ccdf3[-30:])

fig, ax = plt.subplots(1, 2, figsize=(14, 5))

def fitted_exp2(x):
    return exp_func(x, *popt_exp2)
def fitted_pow2(x):
    return power_func(x, *popt_pow2)
def fitted_exp3(x):
    return exp_func(x, *popt_exp3)
def fitted_pow3(x):
    return power_func(x, *popt_pow3)

ax[0].scatter(bin2[:-1], ccdf2)
ax[0].plot(bin2[:-1], fitted_exp2(bin2[:-1]))
ax[0].plot(bin2[3:-1], fitted_pow2(bin2[3:-1]))
ax[0].set_yscale('log')
ax[0].set_xscale('log')

ax[1].scatter(bin3[:-1], ccdf3)
ax[1].plot(bin3[:-1], fitted_exp3(bin3[:-1]))
ax[1].plot(bin3[3:-1], fitted_pow3(bin3[3:-1]))
ax[1].set_yscale('log')
ax[1].set_xscale('log')
#ax[1].set_ylim(1e-4, 1)


```

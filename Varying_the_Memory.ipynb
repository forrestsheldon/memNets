```python
import numpy as np
import scipy as sp
import scipy.sparse as sparse
import matplotlib.pyplot as plt
import generate_adjacency_matrix as gen_adj
import resistor_networks as rnets
from __future__ import division
```

We consider switching between two values of the resistance, $R$ and $(1-\delta)R$. We would like to see what result varying $\delta$ has on the switching time distribution.  As $\delta\to 0$ we get the non-interacting/no-memory picture we have considered thusfar. The $\delta=0$ case is simple in that we do not need to update our resistances at each switching and thus the switching times may be calculated from the initial voltage distribution.  In the case of nonzero $\delta$ we will need to work a bit harder.

## $\delta=0$

```python
# Initialize a network
L = 512
p = 0.6
random_lattice = gen_adj.cubic_2d_random((L,L), p)
v_external = np.empty(L*L) * np.nan
v_external[0:L] = 1
v_external[-L:] = 0
rnet = rnets.ResistorNetworkCC(random_lattice, v_external)

#Solve for the voltages
rnet.solve_voltages('spsolve')
```

```python
def calculate_voltage_drops(G, voltages):
    rows, cols = G.nonzero()
    drops_list = []
    for i, j in zip(rows, cols):
        if i < j:
            drop = np.abs(voltages[i] - voltages[j])
            if drop > 1e-12:
                drops_list.append(drop)
    return drops_list

#get the voltage drops across this network and take a look
voltage_drops = calculate_voltage_drops(rnet.G, rnet.voltages)
hist, bins = np.histogram(np.log10(voltage_drops), 150)

fig, ax = plt.subplots(figsize=(12, 12))
ax.scatter(10**bins[1:], hist)
#ax.set_yscale('log')
ax.set_xscale('log')
#ax.set_ylim(1, hist.max()*2)
ax.set_xlim((10**bins[1:].min())/2, 2*(10**bins[1:].max()))
```

```python
#Calculate the switching time corresponding to each voltage drop
L = 1.
switching_times = L / np.array(voltage_drops)
switching_times.sort()

# The residence times are the gaps between switchings I apply a low end cuttoff to eliminate 0
# residence times from our
residence_times = np.ediff1d(switching_times)
residence_times = residence_times[residence_times > 1e-20]
residence_times.sort()
hist_switch, bins_switch = np.histogram(np.log10(switching_times), 300)
hist, bins = np.histogram(np.log10(residence_times[:-75]), 300)

pdf = hist / np.sum(hist)
cdf = np.cumsum(pdf)
ccdf = 1 - cdf

pdf_switch = hist_switch / np.sum(hist_switch)

fig, ax = plt.subplots(figsize = (15,12))
ax.scatter(10**bins[1:], ccdf)
ax.scatter(10**bins[1:], pdf, c='r')
ax.scatter(10**bins[1:], pdf_switch, c='g')
ax.set_xscale('log')
ax.set_yscale('log')
ax.set_xlim((10**bins[1])/1.5 , 10**bins_switch[-1])
ax.set_ylim(1e-5, np.max(ccdf)*1.5)
```

### $\delta\neq 0$

```python
#initialize the network
L = 64
p = 0.6
R_OFF = 1
delta = 0.1
R_ON = (1 - delta) * R_OFF
G_OFF = 1 / R_OFF
G_ON = 1 / R_ON
L_max = 1
random_lattice = G_OFF * gen_adj.cubic_2d_random((L,L), p)
v_external = np.empty(L*L) * np.nan
v_external[0:L] = 1
v_external[-L:] = 0
rnet = rnets.ResistorNetworkCC(random_lattice, v_external)
```

```python
# initialize filament length matrix
filaments = random_lattice.copy() * (1e-10 / G_OFF)
#initialize voltage drop matrix
voltage_drops = random_lattice.copy()

residence_times = []
switching_indices = []
```

```python
#Solve for the voltages
rnet.solve_voltages('spsolve')
```

```python
# find the shortest switching time in the network
rows, cols = voltage_drops.nonzero()
shortest_switching_time = 1e20
shortest_indices = []
for i, j in zip(rows, cols):
    #matrices are symmetric. consider each memristor once
    if i < j:
        # If the filament has switched, ignore it
        if filaments[i, j] == L_max:
            voltage_drops[i, j] = 0
            voltage_drops[j, i] = 0
        else:
            #Calculate and set the voltage drop
            drop = np.abs(rnet.voltages[i] - rnet.voltages[j])
            if drop >= 1e-12:
                voltage_drops[i, j] = drop
                voltage_drops[j, i] = drop
                switching_time = (L_max - filaments[i, j]) / drop
            else:
                switching_time = 1e21
                voltage_drops[i, j] = 0
                voltage_drops[j, i] = 0
            #If it is shorter than what is stored, set it and drop the indices we have accumulated
            if switching_time < shortest_switching_time:
                shortest_switching_time = switching_time
                shortest_indices = [(i, j)]
            # if it matches the shortest so far, append the indices.
            elif switching_time == shortest_switching_time:
                shortest_indices.append((i, j))
```

```python
#Update the network
filaments = filaments + shortest_switching_time * voltage_drops
for i, j in shortest_indices:
    filaments[i, j] = L_max
    filaments[j, i] = L_max
    rnet.G[i, j] = G_ON
    rnet.G[j, i] = G_ON
    voltage_drops[i, j] = 0
    voltage_drops[j, i] = 0

voltage_drops.eliminate_zeros()
    
residence_times.append(shortest_switching_time)
switching_indices.append(shortest_indices)
```

### Putting it all together

```python
%%timeit

#initialize the network
L = 8
p = 0.6
R_OFF = 1
delta = 0.1
R_ON = (1 - delta) * R_OFF
G_OFF = 1 / R_OFF
G_ON = 1 / R_ON
L_max = 1
random_lattice = G_OFF * gen_adj.cubic_2d_random((L,L), p)
v_external = np.empty(L*L) * np.nan
v_external[0:L] = 1
v_external[-L:] = 0
rnet = rnets.ResistorNetworkCC(random_lattice, v_external)

# initialize filament length matrix
filaments = random_lattice.copy() * (1e-10 / G_OFF)
#initialize voltage drop matrix
voltage_drops = random_lattice.copy()

#initialize lists for saving results
residence_times = []
switching_indices = []

while voltage_drops.max() > 0:
    
    #Solve for the voltages
    rnet.solve_voltages('spsolve')
    
    # find the shortest switching time in the network
    rows, cols = voltage_drops.nonzero()
    shortest_switching_time = 1e20
    shortest_indices = []
    for i, j in zip(rows, cols):
        #matrices are symmetric. consider each memristor once
        if i < j:
            # If the filament has switched, ignore it
            if filaments[i, j] == L_max:
                voltage_drops[i, j] = 0
                voltage_drops[j, i] = 0
            else:
                #Calculate and set the voltage drop
                drop = np.abs(rnet.voltages[i] - rnet.voltages[j])
                if drop >= 1e-12:
                    voltage_drops[i, j] = drop
                    voltage_drops[j, i] = drop
                    switching_time = (L_max - filaments[i, j]) / drop
                else:
                    switching_time = 1e21
                    voltage_drops[i, j] = 0
                    voltage_drops[j, i] = 0
                #If it is shorter than what is stored, set it and drop the indices we have accumulated
                if switching_time < shortest_switching_time:
                    shortest_switching_time = switching_time
                    shortest_indices = [(i, j)]
                # if it matches the shortest so far, append the indices.
                elif switching_time == shortest_switching_time:
                    shortest_indices.append((i, j))

    #Update the network
    filaments = filaments + shortest_switching_time * voltage_drops
    for i, j in shortest_indices:
        filaments[i, j] = L_max
        filaments[j, i] = L_max
        rnet.G[i, j] = G_ON
        rnet.G[j, i] = G_ON
        voltage_drops[i, j] = 0
        voltage_drops[j, i] = 0
    
    #eliminate zero entries from storage
    voltage_drops.eliminate_zeros()
    
    residence_times.append(shortest_switching_time)
    switching_indices.append(shortest_indices)
```

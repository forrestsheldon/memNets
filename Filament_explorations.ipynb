### The Model

```python
def G_hard_switching(filament_length, junction_length, G_ON=1., G_OFF=1e-2):
    if filament_length < junction_length:
        return G_OFF
    else:
        return G_ON
    
def deriv_filament_length_dt_growth(t, filament_length, junction_length, V, G_ON=1., G_OFF=1e-2):
    alpha = 1.5
    return alpha * V

def deriv_filament_length_dt_burn(t, filament_length, junction_length, V, G_ON=1., G_OFF=1e-2):
    V_thresh = .5
    if filament_length == junction_length and V > V_thresh:
        return -1e20
    else:
        return deriv_filament_length_dt_growth(t, filament_length, junction_length, V, G_ON=1., G_OFF=1e-2)
    
def deriv_filament_length_dt_relax(t, filament_length, junction_length, V):
    beta = 0.5
    noise = 1e-3
    return deriv_filament_length_dt_growth(t, filament_length, junction_length, V) - beta * (junction_length - filament_length)

def deriv_filament_length_dt_all(t, filament_length, junction_length, V):
    V_thresh = .5
    if filament_length == junction_length and V > V_thresh:
        return -1e20
    else:
        return deriv_filament_length_dt_relax(t, filament_length, junction_length, V)

def deriv_filament_length_dt_all_noisy(t, filament_length, junction_length, V):
    V_thresh = .5
    noise_scale = 1e-2
    if filament_length == junction_length and V > V_thresh:
        return -1e20
    else:
        return deriv_filament_length_dt_relax(t, filament_length, junction_length, V) + noise_scale * np.random.randn()
```

### The Network

```python
import generate_adjacency_matrix as gen_adj
import memristor_networks as mnets
from IPython.display import clear_output
import sys
```

```python
filament_zero = 1e-12
N = 10
lattice_shape = (N, N)
num_nodes = lattice_shape[0]*lattice_shape[1]

def exp_decay(x):
    return 0.7 * np.exp(-1.5*(x-1))

initial_filaments = filament_zero * gen_adj.locally_connected(lattice_shape, exp_decay)
```

### Current Memristor Model

```python
def conductance_function(filament_length):
    return G_hard_switching(filament_length, 1)

def xdot(t, x, V, junction_length):
    return deriv_filament_length_dt_all(t, x, junction_length, V)
```

### External Voltage

```python
pts_per_second = 1000
t_final = 4
time = np.linspace(0, t_final, t_final * pts_per_second)
V_peak = 3
V = V_peak * np.ones_like(time)
v_external = np.empty((num_nodes, t_final * pts_per_second)) * np.nan
v_external[0:N, :] = V
v_external[-1*N:, :] = 0
```

```python
asnet = mnets.MemristorNetwork(initial_filaments.copy(), v_external[:,0], conductance_function, xdot)
asnet.solve_voltages('cg')
fig, ax = plt.subplots(figsize=(10,10))
asnet.display_grid(ax, lattice_shape, 'voltage')
```

```python
current_all = []


for t_idx, t in enumerate(time[:-1]):
    clear_output()
    print t / t_final
    sys.stdout.flush()
    # Update the external voltages
    asnet.external_voltages = v_external[:, t_idx]
    # solve for the voltages in the network
    asnet.solve_voltages('cg')
    # Find the sum of the external current flowing INTO the network
    I_external = asnet.external_current()
    current_all.append(I_external[I_external>0].sum())
    # Now update the conductances
    dt = time[t_idx+1] - t
    asnet.integrate_state_variables(dt, method='myrk4')
    
clear_output()
print 1
sys.stdout.flush()
# Update the external voltages
asnet.external_voltages = v_external[:, t_idx+1]
# solve for the voltages in the network
asnet.solve_voltages('cg')
# Find the sum of the external current flowing INTO the network
I_external = asnet.external_current()
current_all.append(I_external[I_external>0].sum())
```

```python
fig, ax = plt.subplots(figsize=(25,8))
ax.plot(time, current_all)
```

```python
def residence_times_cond(current, voltage, t, cutoff=0, tolerance=1e-5):
    """
    Returns a list of time intervals during which the conductance of the device was constant within a given tolerance.
    A cutoff may be introduced and should be at least the temporal spacing.
    """
    def conductance(index):
        if voltage[index] != 0:
            return current[index] / voltage[index]
        else:
            return np.nan
    
    res_times = []
    beginnings = []
    endings = []
    beginning = 0

    for idx, time in enumerate(t[:-1]):
        
        # a new conductance state is initialized if the conductance goes from nan to a new value
        if np.isnan(conductance(idx-1)) and not np.isnan(conductance(idx)):
            beginning = idx
        
        # We exit the state if the conductance changes to a value further away than the tolerance
        # or if it becomes nan
        if np.abs(conductance(idx+1) - conductance(beginning)) > tolerance or \
        (np.isnan(conductance(idx+1)) and not np.isnan(conductance(idx))):
            state_time = t[idx] - t[beginning]
            # only include it if the length of the state is longer than the cutoff.
            if state_time > cutoff:
                res_times.append(state_time)
                beginnings.append(beginning)
                endings.append(idx)
            beginning = idx+1
    
    return res_times, beginnings, endings
```

```python
res_all, beg, end = residence_times_cond(current_all, V, time)
res_all.sort()
hist, bins = np.histogram(res_all[:-16], bins=20)
fig, ax = plt.subplots()
ax.plot(bins[:-1], hist)
```

```python
fig, ax = plt.subplots()
ax.scatter(bins[:-1], hist)
ax.set_yscale('log')
```

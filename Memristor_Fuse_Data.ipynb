# Memristor Fuse Data

```python
%matplotlib inline
```

```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
```

```python
L128_p6_ON1_1 = np.load('memristor_fuse_data/FuseNetwork_L128_p6_ON1.1run1.npz')
avalanches_L128_p6_ON1_1 = np.array([np.sum(i) for i in L128_p6_ON1_1['avalanches']])
L128_p6_ON1_5 = np.load('memristor_fuse_data/FuseNetwork_L128_p6_ON1.5run1.npz')
avalanches_L128_p6_ON1_5 = np.array([np.sum(i) for i in L128_p6_ON1_5['avalanches']])
L128_p6_ON2 = np.load('memristor_fuse_data/FuseNetwork_L128_p6_ON2run1.npz')
avalanches_L128_p6_ON2 = np.array([np.sum(i) for i in L128_p6_ON2['avalanches']])
L128_p6_ON4 = np.load('memristor_fuse_data/FuseNetwork_L128_p6_ON4run1.npz')
avalanches_L128_p6_ON4 = np.array([np.sum(i) for i in L128_p6_ON4['avalanches']])
L128_p6_ON8 = np.load('memristor_fuse_data/FuseNetwork_L128_p6_ON8run1.npz')
avalanches_L128_p6_ON8 = np.array([np.sum(i) for i in L128_p6_ON8['avalanches']])
L128_p6_ON16 = np.load('memristor_fuse_data/FuseNetwork_L128_p6_ON16run1.npz')
avalanches_L128_p6_ON16 = np.array([np.sum(i) for i in L128_p6_ON16['avalanches']])
L128_p6_ON100 = np.load('memristor_fuse_data/FuseNetwork_L128_p6_ON100run1.npz')
avalanches_L128_p6_ON100 = np.array([np.sum(i) for i in L128_p6_ON100['avalanches']])
```

```python
fig, ax = plt.subplots(7, 1, figsize=(15, 21))
ax[0].plot(avalanches_L128_p6_ON1_1)
ax[0].set_title('ON=1.1')
ax[1].plot(avalanches_L128_p6_ON1_5)
ax[1].set_title('ON=1.5')
ax[2].plot(avalanches_L128_p6_ON2)
ax[2].set_title('ON=2')
ax[3].plot(avalanches_L128_p6_ON4)
ax[3].set_title('ON=4')
ax[4].plot(avalanches_L128_p6_ON8)
ax[4].set_title('ON=8')
ax[5].plot(avalanches_L128_p6_ON16)
ax[5].set_title('ON=16')
ax[6].plot(avalanches_L128_p6_ON100)
ax[6].set_title('ON=100')
for a in ax:
    a.set_xlim(0, 10500)
    a.set_ylim(0,600)

fig.tight_layout()
```

```python
hist_L128_p6_ON1_1, bins_L128_p6_ON1_1 = np.histogram(np.log10(avalanches_L128_p6_ON1_1[
                                         np.logical_and(avalanches_L128_p6_ON1_1>0, avalanches_L128_p6_ON1_1<100)]), bins=30)
hist_L128_p6_ON1_5, bins_L128_p6_ON1_5 = np.histogram(np.log10(avalanches_L128_p6_ON1_5[
                                         np.logical_and(avalanches_L128_p6_ON1_5>0, avalanches_L128_p6_ON1_5<100)]), bins=30)
hist_L128_p6_ON2, bins_L128_p6_ON2 = np.histogram(np.log10(avalanches_L128_p6_ON2[
                                         np.logical_and(avalanches_L128_p6_ON2>0, avalanches_L128_p6_ON2<100)]), bins=30)
hist_L128_p6_ON4, bins_L128_p6_ON4 = np.histogram(np.log10(avalanches_L128_p6_ON4[
                                         np.logical_and(avalanches_L128_p6_ON4>0, avalanches_L128_p6_ON4<200)]), bins=30)
hist_L128_p6_ON8, bins_L128_p6_ON8 = np.histogram(np.log10(avalanches_L128_p6_ON8[
                                         np.logical_and(avalanches_L128_p6_ON8>0, avalanches_L128_p6_ON8<200)]), bins=30)
hist_L128_p6_ON16, bins_L128_p6_ON16 = np.histogram(np.log10(avalanches_L128_p6_ON16[
                                         np.logical_and(avalanches_L128_p6_ON16>0, avalanches_L128_p6_ON16<200)]), bins=30)
hist_L128_p6_ON100, bins_L128_p6_ON100 = np.histogram(np.log10(avalanches_L128_p6_ON100[
                                         np.logical_and(avalanches_L128_p6_ON100>0, avalanches_L128_p6_ON100<200)]), bins=30)

fig, ax = plt.subplots(7, 1, figsize=(8, 28))

ax[0].scatter(10**bins_L128_p6_ON1_1[1:], hist_L128_p6_ON1_1, c='b')
ax[0].set_title('ON=1.1')
ax[1].scatter(10**bins_L128_p6_ON1_5[1:], hist_L128_p6_ON1_5, c='r')
ax[1].set_title('ON=1.5')
ax[2].scatter(10**bins_L128_p6_ON2[1:], hist_L128_p6_ON2, c='g')
ax[2].set_title('ON=2')
ax[3].scatter(10**bins_L128_p6_ON4[1:], hist_L128_p6_ON4, c='c')
ax[3].set_title('ON=4')
ax[4].scatter(10**bins_L128_p6_ON8[1:], hist_L128_p6_ON8, c='c')
ax[4].set_title('ON=8')
ax[5].scatter(10**bins_L128_p6_ON16[1:], hist_L128_p6_ON16, c='c')
ax[5].set_title('ON=16')
ax[6].scatter(10**bins_L128_p6_ON100[1:], hist_L128_p6_ON100, c='c')
ax[6].set_title('ON=100')
for a in ax:
 a.set_xscale('log')
 a.set_yscale('log')
 a.set_xlim(1.5, 200)
 a.set_ylim(1, 1000)
    
fig.tight_layout()
```

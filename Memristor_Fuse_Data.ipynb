```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
```

```python
L128_p6_ON1_1 = np.load('memristor_fuse_data/FuseNetwork_L128_p6_ON1.1run1.npz')
avalanches_L128_p6_ON1_1 = np.array([np.sum(i) for i in L128_p6_ON1_1['avalanches']])
steps_L128_p6_ON1_1 = np.array([len(i) for i in L128_p6_ON1_1['avalanches']])
L128_p6_ON1_5 = np.load('memristor_fuse_data/FuseNetwork_L128_p6_ON1.5run1.npz')
avalanches_L128_p6_ON1_5 = np.array([np.sum(i) for i in L128_p6_ON1_5['avalanches']])
steps_L128_p6_ON1_5 = np.array([len(i) for i in L128_p6_ON1_5['avalanches']])
L128_p6_ON2 = np.load('memristor_fuse_data/FuseNetwork_L128_p6_ON2run1.npz')
avalanches_L128_p6_ON2 = np.array([np.sum(i) for i in L128_p6_ON2['avalanches']])
steps_L128_p6_ON2 = np.array([len(i) for i in L128_p6_ON2['avalanches']])
L128_p6_ON4 = np.load('memristor_fuse_data/FuseNetwork_L128_p6_ON4run1.npz')
avalanches_L128_p6_ON4 = np.array([np.sum(i) for i in L128_p6_ON4['avalanches']])
steps_L128_p6_ON4 = np.array([len(i) for i in L128_p6_ON4['avalanches']])
L128_p6_ON8 = np.load('memristor_fuse_data/FuseNetwork_L128_p6_ON8run1.npz')
avalanches_L128_p6_ON8 = np.array([np.sum(i) for i in L128_p6_ON8['avalanches']])
steps_L128_p6_ON8 = np.array([len(i) for i in L128_p6_ON8['avalanches']])
L128_p6_ON16 = np.load('memristor_fuse_data/FuseNetwork_L128_p6_ON16run1.npz')
avalanches_L128_p6_ON16 = np.array([np.sum(i) for i in L128_p6_ON16['avalanches']])
steps_L128_p6_ON16 = np.array([len(i) for i in L128_p6_ON16['avalanches']])
L128_p6_ON100 = np.load('memristor_fuse_data/FuseNetwork_L128_p6_ON100run1.npz')
avalanches_L128_p6_ON100 = np.array([np.sum(i) for i in L128_p6_ON100['avalanches']])
steps_L128_p6_ON100 = np.array([len(i) for i in L128_p6_ON100['avalanches']])
```

```python
fig, ax = plt.subplots(7, 2, figsize=(15, 21))

ax[0,0].plot(avalanches_L128_p6_ON1_1)
ax[0,0].plot(np.cumsum(avalanches_L128_p6_ON1_1) * 100 / np.sum(avalanches_L128_p6_ON1_1))
ax[0,0].set_title('ON=1.1')
ax[0,1].plot(steps_L128_p6_ON1_1)

ax[1,0].plot(avalanches_L128_p6_ON1_5)
ax[1,0].plot(np.cumsum(avalanches_L128_p6_ON1_5) * 100 / np.sum(avalanches_L128_p6_ON1_5))
ax[1,0].set_title('ON=1.5')
ax[1,1].plot(steps_L128_p6_ON1_5)

ax[2,0].plot(avalanches_L128_p6_ON2)
ax[2,0].plot(np.cumsum(avalanches_L128_p6_ON2) * 100 / np.sum(avalanches_L128_p6_ON2))
ax[2,0].set_title('ON=2')
ax[2,1].plot(steps_L128_p6_ON2)

ax[3,0].plot(avalanches_L128_p6_ON4)
ax[3,0].plot(np.cumsum(avalanches_L128_p6_ON4) * 100 / np.sum(avalanches_L128_p6_ON4))
ax[3,0].set_title('ON=4')
ax[3,1].plot(steps_L128_p6_ON4)

ax[4,0].plot(avalanches_L128_p6_ON8)
ax[4,0].plot(np.cumsum(avalanches_L128_p6_ON8) * 100 / np.sum(avalanches_L128_p6_ON8))
ax[4,0].set_title('ON=8')
ax[4,1].plot(steps_L128_p6_ON8)

ax[5,0].plot(avalanches_L128_p6_ON16)
ax[5,0].plot(np.cumsum(avalanches_L128_p6_ON16) * 100 / np.sum(avalanches_L128_p6_ON16))
ax[5,0].set_title('ON=16')
ax[5,1].plot(steps_L128_p6_ON16)

ax[6,0].plot(avalanches_L128_p6_ON100)
ax[6,0].plot(np.cumsum(avalanches_L128_p6_ON100) * 100 / np.sum(avalanches_L128_p6_ON100))
ax[6,0].set_title('ON=100')
ax[6,1].plot(steps_L128_p6_ON100)


#for a in ax:
#    a.set_xlim(0, 10500)
#    a.set_ylim(0,600)

fig.tight_layout()
```

```python
#avalanches_L128_p6_ON1_1 = avalanches_L128_p6_ON1_1[1000:5000]
hist_L128_p6_ON1_1, bins_L128_p6_ON1_1 = np.histogram(avalanches_L128_p6_ON1_1[
                                         np.logical_and(avalanches_L128_p6_ON1_1>0, avalanches_L128_p6_ON1_1<100)], bins=30)
hist_steps_L128_p6_ON1_1, bins_steps_L128_p6_ON1_1 = np.histogram(steps_L128_p6_ON1_1[avalanches_L128_p6_ON1_1>0], bins=10)

#avalanches_L128_p6_ON1_5 = avalanches_L128_p6_ON1_5[4000:8000]
hist_L128_p6_ON1_5, bins_L128_p6_ON1_5 = np.histogram(avalanches_L128_p6_ON1_5[
                                         np.logical_and(avalanches_L128_p6_ON1_5>0, avalanches_L128_p6_ON1_5<100)], bins=30)
hist_steps_L128_p6_ON1_5, bins_steps_L128_p6_ON1_5 = np.histogram(steps_L128_p6_ON1_5[avalanches_L128_p6_ON1_5>0], bins=12)

#avalanches_L128_p6_ON2 = avalanches_L128_p6_ON2[100:4000]
hist_L128_p6_ON2, bins_L128_p6_ON2 = np.histogram(avalanches_L128_p6_ON2[
                                         np.logical_and(avalanches_L128_p6_ON2>0, avalanches_L128_p6_ON2<100)], bins=30)
hist_steps_L128_p6_ON2, bins_steps_L128_p6_ON2 = np.histogram(steps_L128_p6_ON2[avalanches_L128_p6_ON2>0], bins=12)

#avalanches_L128_p6_ON4 = avalanches_L128_p6_ON4[100:4000]
hist_L128_p6_ON4, bins_L128_p6_ON4 = np.histogram(avalanches_L128_p6_ON4[
                                         np.logical_and(avalanches_L128_p6_ON4>0, avalanches_L128_p6_ON4<200)], bins=30)
hist_steps_L128_p6_ON4, bins_steps_L128_p6_ON4 = np.histogram(steps_L128_p6_ON4[avalanches_L128_p6_ON4>0], bins=20)

#avalanches_L128_p6_ON8 = avalanches_L128_p6_ON8[400:1900]
hist_L128_p6_ON8, bins_L128_p6_ON8 = np.histogram(avalanches_L128_p6_ON8[
                                         np.logical_and(avalanches_L128_p6_ON8>0, avalanches_L128_p6_ON8<200)], bins=30)
hist_steps_L128_p6_ON8, bins_steps_L128_p6_ON8 = np.histogram(steps_L128_p6_ON8[avalanches_L128_p6_ON8>0], bins=30)

#avalanches_L128_p6_ON16 = avalanches_L128_p6_ON16[1000:2300]
hist_L128_p6_ON16, bins_L128_p6_ON16 = np.histogram(avalanches_L128_p6_ON16[
                                         np.logical_and(avalanches_L128_p6_ON16>0, avalanches_L128_p6_ON16<200)], bins=30)
hist_steps_L128_p6_ON16, bins_steps_L128_p6_ON16 = np.histogram(steps_L128_p6_ON16[avalanches_L128_p6_ON16>0], bins=30)

#avalanches_L128_p6_ON100 = avalanches_L128_p6_ON100[400:1900]
hist_L128_p6_ON100, bins_L128_p6_ON100 = np.histogram(avalanches_L128_p6_ON100[
                                         np.logical_and(avalanches_L128_p6_ON100>0, avalanches_L128_p6_ON100<200)], bins=30)
hist_steps_L128_p6_ON100, bins_steps_L128_p6_ON100 = np.histogram(steps_L128_p6_ON100[avalanches_L128_p6_ON100>0], bins=30)

fig, ax = plt.subplots(7, 2, figsize=(15, 28))

ax[0,0].scatter(bins_L128_p6_ON1_1[1:], hist_L128_p6_ON1_1, c='b')
ax[0,0].set_title('ON=1.1')
ax[0,0].set_xlim(1, 200)
ax[0,0].set_ylim(1, 1500)

ax[1,0].scatter(10**bins_L128_p6_ON1_5[1:], hist_L128_p6_ON1_5, c='b')
ax[1,0].set_title('ON=1.5')
ax[1,0].set_xlim(1, 200)
ax[1,0].set_ylim(1, 1500)

ax[2,0].scatter(10**bins_L128_p6_ON2[1:], hist_L128_p6_ON2, c='b')
ax[2,0].set_title('ON=2')
ax[2,0].set_xlim(1, 200)
ax[2,0].set_ylim(1, 1500)

ax[3,0].scatter(10**bins_L128_p6_ON4[1:], hist_L128_p6_ON4, c='b')
ax[3,0].set_title('ON=4')
ax[3,0].set_xlim(1, 200)
ax[3,0].set_ylim(1, 1500)

ax[4,0].scatter(10**bins_L128_p6_ON8[1:], hist_L128_p6_ON8, c='b')
ax[4,0].set_title('ON=8')
ax[4,0].set_xlim(1, 200)
ax[4,0].set_ylim(1, 1500)

ax[5,0].scatter(10**bins_L128_p6_ON16[1:], hist_L128_p6_ON16, c='b')
ax[5,0].set_title('ON=16')
ax[5,0].set_xlim(1, 200)
ax[5,0].set_ylim(1, 1500)

ax[6,0].scatter(10**bins_L128_p6_ON100[1:], hist_L128_p6_ON100, c='b')
ax[6,0].set_title('ON=100')
ax[6,0].set_xlim(1, 200)
ax[6,0].set_ylim(1, 1500)


for plot_row in ax:
    for a in plot_row:
        a.set_xscale('log')
        a.set_yscale('log')

    
fig.tight_layout()
```

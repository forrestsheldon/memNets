# Memristor Fuse Data

```python
%matplotlib inline
```

```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
```

```python
L128_p5_ON10 = np.load('memristor_fuse_data/FuseNetwork_L128_p5_ON10run1.npz')
avalanches_p5_ON10 = np.array([np.sum(i) for i in L128_p5_ON10['avalanches']])
L128_p5_ON100 = np.load('memristor_fuse_data/FuseNetwork_L128_p5_ON100run1.npz')
avalanches_p5_ON100 = np.array([np.sum(i) for i in L128_p5_ON100['avalanches']])
L128_p5_ON1000 = np.load('memristor_fuse_data/FuseNetwork_L128_p5_ON1000run1.npz')
avalanches_p5_ON1000 = np.array([np.sum(i) for i in L128_p5_ON1000['avalanches']])
L128_p5_ON10000 = np.load('memristor_fuse_data/FuseNetwork_L128_p5_ON10000run1.npz')
avalanches_p5_ON10000 = np.array([np.sum(i) for i in L128_p5_ON10000['avalanches']])
```

```python
fig, ax = plt.subplots(4, 1, figsize=(15, 12))
ax[0].plot(avalanches_p5_ON10)
ax[1].plot(avalanches_p5_ON100)
ax[2].plot(avalanches_p5_ON1000)
ax[3].plot(avalanches_p5_ON10000)
```

```python
hist_p5_ON10, bins_p5_ON10 = np.histogram(np.log10(avalanches_p5_ON10[np.logical_and(avalanches_p5_ON10>0, avalanches_p5_ON10<1500)]), bins=30)
hist_p5_ON100, bins_p5_ON100 = np.histogram(np.log10(avalanches_p5_ON100[np.logical_and(avalanches_p5_ON100>0, avalanches_p5_ON100<1500)]), bins=30)
hist_p5_ON1000, bins_p5_ON1000 = np.histogram(np.log10(avalanches_p5_ON1000[np.logical_and(avalanches_p5_ON1000>0, avalanches_p5_ON1000<1500)]), bins=30)
hist_p5_ON10000, bins_p5_ON10000 = np.histogram(np.log10(avalanches_p5_ON10000[np.logical_and(avalanches_p5_ON10000>0, avalanches_p5_ON10000<1500)]), bins=30)

fig, ax = plt.subplots(4, 1, figsize=(8, 20))

ax[0].scatter(10**bins_p5_ON10[1:], hist_p5_ON10, c='b')
ax[1].scatter(10**bins_p5_ON100[1:], hist_p5_ON100, c='r')
ax[2].scatter(10**bins_p5_ON1000[1:], hist_p5_ON1000, c='g')
ax[3].scatter(10**bins_p5_ON10000[1:], hist_p5_ON10000, c='c')
for a in ax:
 a.set_xscale('log')
 a.set_yscale('log')
 a.set_xlim(1.5, 900)
 a.set_ylim(1.5, 200)
```

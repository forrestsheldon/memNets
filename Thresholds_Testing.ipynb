# Testing for Threshold Models

```python
%matplotlib inline
```

Here I would like to run testing to get some numerics ready for threshold models.  The idea is as follows:  We set up a network with a particular voltage.  AS will switch until the none carry a current above the threshold.  At this point, we increase the voltage slightly, enough to cause the next memristor to switch. This will increase the current and likely cause other memristors to switch down the line.  When all memristors have switched, we collect the number of memristors that switch for each increase and bin these to hopefully observe a power law.

### Setting the threshold

I need to know what threshold to apply for the memristors.  To that end, I need to look at random resistor networks voltage distributions and decide where would be sensible to put it.

```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
import matplotlib as mpl
import generate_adjacency_matrix as gen_adj
import resistor_networks as rnets
from __future__ import division
```

```python
L = 128
p = 0.6
random_lattice = gen_adj.cubic_2d_random((L,L), p)
v_external = np.empty(L*L) * np.nan
voltage = 1
v_external[0:L] = voltage
v_external[-L:] = 0

rnet = rnets.ResistorNetworkCC(random_lattice, v_external)
rnet.solve_voltages('spsolve')

#fig, ax = plt.subplots(figsize=(10,10))
#rnet.display_grid(ax, (L,L), 'voltage')
rnet.conductivity()
```

```python
def calculate_voltage_drops(G, voltages):
    rows, cols = G.nonzero()
    drops_list = []
    for i, j in zip(rows, cols):
        if i < j:
            drop = np.abs(voltages[i] - voltages[j])
            if drop > 1e-12:
                drops_list.append(drop)
    return drops_list
```

```python
voltage_drops = calculate_voltage_drops(rnet.G, rnet.voltages)
len(voltage_drops)
```

```python
hist, bins = np.histogram(np.log10(voltage_drops), 150)

fig, ax = plt.subplots(figsize=(12, 12))
ax.scatter(10**bins[1:], hist)
#ax.set_yscale('log')
ax.set_xscale('log')
#ax.set_ylim(1, hist.max()*2)
ax.set_xlim((10**bins[1:].min())/2, 2*(10**bins[1:].max()))
```

So a threshold of 0.1 or 0.05 would leave most of the memristors unswitched.  To confirm, I'll switch all memristors above a threshold repeatedly until none others want to switch.  Then I'll look at the voltages of bonds below that threshold.  I think a R_ON value of something like 0.01 might be appropriate althrough I should check at 0.001 and so on to confirm.

```python
def switch_memristors(G, voltages, I_thresh, G_ON=100., G_OFF=1.):
    rows, cols = G.nonzero()
    switch_counter = 0
    for i, j in zip(rows, cols):
        if i < j:
            current = G[i, j] * np.abs(voltages[i] - voltages[j])
            if G[i, j] == G_OFF and current >= I_thresh:
                G[i, j] = G_ON
                G[j, i] = G_ON
                switch_counter += 1
    return G, switch_counter

def find_maximum_unswitched_voltage(G, voltages, G_ON=100., G_OFF=1.):
    rows, cols = G.nonzero()
    max_unswitched_voltage = 0
    for i, j in zip(rows, cols):
        if G[i, j] == G_OFF and i < j:
            drop = np.abs(voltages[i] - voltages[j])
            if drop > max_unswitched_voltage:
                max_unswitched_voltage = drop
                indices = (i, j)
    return max_unswitched_voltage, indices

def find_maximum_switched_voltage(G, voltages, G_ON=100., G_OFF=1.):
    rows, cols = G.nonzero()
    max_switched_voltage = 0
    for i, j in zip(rows, cols):
        if G[i, j] == G_ON and i < j:
            drop = np.abs(voltages[i] - voltages[j])
            if drop > max_switched_voltage:
                max_switched_voltage = drop
                indices = (i, j)
    return max_switched_voltage, indices
```

```python
I_threshold = 0.02
G_OFF = 1
G_ON = 100
max_unswitched_voltage, indices = find_maximum_unswitched_voltage(rnet.G, rnet.voltages)
avalanche_size = []
conductivities = []
conductivities.append(rnet.conductivity())
```

```python
while max_unswitched_voltage != 0:
    counter = 0
    while max_unswitched_voltage*G_OFF > I_threshold:
        rnet.G, num_switched = switch_memristors(rnet.G, rnet.voltages, I_threshold)
        counter += num_switched
    
        rnet.solve_voltages('spsolve')
        max_unswitched_voltage, indices = find_maximum_unswitched_voltage(rnet.G, rnet.voltages)
    
    avalanche_size.append(counter)
    conductivities.append(rnet.conductivity())
    voltage = voltage * I_threshold / max_unswitched_voltage + 1e-15
    rnet.external_voltages[0:L] = voltage
    rnet.solve_voltages('spsolve')
    max_unswitched_voltage, indices = find_maximum_unswitched_voltage(rnet.G, rnet.voltages)
```

### The Self Consistency Equation

To form a mean field theory for the memristor networks, we consider the power dissipated,
$$P = \sum_i \sigma_i \Delta V_i^2.$$
We need some way to approximate $\Delta V$ in terms of the average network conductivity.  To do so, we write the total network conductivity in terms of effective medium theory,
$$P = G(\{\sigma_i\})V^2 \approx G(f(\phi))V^2 $$
where $G(\{\sigma_i\})$ is the exact conductance which is some complicated function of the full set of conductances $\{\sigma_i\}$ and $G(f(\phi))=G$ is the approximate conductance as given by effective medium theory where $f$ is the fraction of conductors in the $ON$ state and $\phi = \frac{\sum_i \sigma_i}{N}$.  To get the local power back we insert factors of $\phi$ back in,
$$ P = \sum_i \sigma_i \frac{G(f(\phi)))V^2}{N\phi}.$$
In this form, we recognize the local voltage, $\sqrt{\frac{G(f(\phi))}{\phi}} \frac{V}{\sqrt{N}} = h(\phi) v$.  When this exceeds a threshold $t_i$, the conductance of the link transitions from $\sigma_i = g_{OFF}$ to $g_{ON}$.  In order for this to be consistent with the effective medium theory, we must have
$$\phi = g_{OFF} P(t_i > h(\phi) v) + g_{ON} P(t_i < h(\phi) v) = g_{OFF} + (g_{ON} - g_{OFF})\int^{h(\phi) v}_0 p(t) dt$$
Note that this can be rewritten in terms of the concentration of resistors in the ON state, $f$ as,
$$f = \int^{H(f) v}_0 p(t) dt$$
where $H(f) = h(\phi(f)) = \sqrt{\frac{G(f)}{\phi(f)}}$
Our mean field theory thus reduces to understanding the function $H(f)$ and it's consequences on the self consistency equation.  To obtain $H(f)$ we appeal to effective medium theory.

### Effective Medium Theory

To obtain the approximate conductance of the network, we attempt to replace it with a uniform medium whose conductance matches that of the disordered medium as closely as possible.  To this end, we consider a lattice of uniform conductance $g_m$.  The resistors will obey the local equation $g_m V_m = I_m$.  Deep in the lattice, we replace a single conductor with one of conductance $g$.  This will perturb the current and voltage around the uniform values, $I_m + \delta I = g(V_m +\delta V)$.  Deep in the network, we do not expect this to alter the total current flowing through the network.  The displaced current must then circulate back through the network.  This perturbs the voltage over the single resistor as, $\delta I = -G \delta V$ where $G$ is the conductance of the lattice.  Symmetry arguments or Green's function methods give $G=g_m$. So, solving
$$I_m + \delta I = g(V_m +\delta V)$$
$$\delta I = -g \delta V$$
$$I_m = g_m V_m$$
for $\delta V$ gives,
$$\delta V = V_m \frac{g_m - g}{g_m + g}.$$
Note that for $g=g_m$, $\delta V=0$ and the sign of $\delta V$ opposes the direction of $g$ so that as the bond becomes superconducting, the voltage drop goes to zero (and as it goes to zero, the voltage drop goes to $2V_m$).  We then require that as we replace $g$ with values pulled from a distibution $p(g)$, the average change in the voltage vanishes, $\langle \delta V \rangle = 0$.  For two useful distributions,
$$p(g) = p \delta(g-1) + (1-p) \delta(g)$$
$$g_m = (2p-1) \quad p\in [\frac{1}{2}, 1]$$
reproducing the critical percolation threshold at $p=\frac{1}{2}$
$$p(g) = p\delta(g - g_{ON}) + (1-p)\delta(g - g_{OFF})$$
$$g_m = \frac{(2p-1)(g_{ON}-g_{OFF}) + \sqrt{(2p-1)^2(g_{ON} - g_{OFF})^2+4g_{ON}g_{OFF}}}{2} $$

Now that the ingredients are assembled, let's work on visualization:

First, I'd like to see the effective medium approximation and the load this creates $H(f)$,

```python
g_ON = 100
g_OFF = 1

def G_effectivemedium(f):
    return ((2*f-1) * (g_ON-g_OFF) + np.sqrt((2*f-1)**2 * (g_ON-g_OFF)**2 + 4*g_ON*g_OFF)) / 2

def f_of_phi(phi):
    return (phi - g_OFF) / (g_ON - g_OFF)

def phi_of_f(f):
    return g_OFF + (g_ON - g_OFF) * f
```

#### Limits

In the limits $g_{on}\to\infty$ we can expand our $G(f)$ to first order in $\alpha = g_{off}/g_{on}$.  This gives
$$G_-(f) = -\frac{g_{on}\alpha}{(2f-1)}\quad 0\le f \le \frac{1}{2}$$
$$G_+(f) = g_{on}(2f-1)\left(1 - \alpha(1 - \frac{1}{(2f-1)^2})\right)$$


```python
def G_minus(f):
    return g_OFF / (1 - 2*f)

def G_plus(f):
    x = 2*f - 1
    return g_ON * (x - g_OFF / g_ON * (x - 1 / x))
```

### Avalanche Size Distribution

To attack the avalanche size distribution, we consider our description of the system in the mean field limit,
$$P = \sum_i \sigma_i \frac{G(f)}{\phi(f)} \frac{V^2}{N}$$
Where $phi = \frac{\sum_i \sigma_i}{N} = g_{OFF} + f(g_{ON}-g_{OFF})$.  For convenience, we will write the voltage across all memristors as $H(f)v = \sqrt{\frac{G(f)}{\phi(f)}}\frac{V}{\sqrt{N}}$. Tripping a single memristor changes $f\to f+\frac{1}{N}$. This then changes the load across all other memristors to $H(f+\frac{1}{N})v \approx H(f)v + \frac{H'(f)v}{N}$.  This process will cause all memristors within the window $\{H(f)v, H(f)v + \frac{H'(f)v}{N}\}$ to turn on which will alter the load again, thus creating a potential avalanche. In the discrete time model we are considering, avalanches will come as a sequence, $Z_1, Z_2, Z_3, ...Z_{n+1}$ with $Z_0 = 1,\; Z_{n+1} = 0$. ### One Dimensional Models

To build some intuition, we begin with some one dimensional models.  First, we consider a chain of $N$ memristors connected in series.  Each has some threshold $t_i$ above which it will transition from $g_{OFF} \to g_{ON}$.  For a current controlled system, clearly we won't see any changes as memristors switch, so we consider a voltage controlled system.  In this case, the total current through every memristor will be $I = G_t V$ with $G_t = \frac{1}{\sum_i \frac{1}{\sigma_i}}$ Where $\sigma_i$ is the conductance of each memristor.  If $n$ of the memristors have switched to $g_{ON}$, $G_T$ can be written as,
$$G_T = \frac{g_{ON}g_{OFF}}{N g_{ON} - n(g_{ON} - g_{OFF})} = \frac{g_{ON}g_{OFF}}{N (g_{ON} - f(g_{ON} - g_{OFF}))}.$$
Averaged over many ensembles, or for large $N$, the fraction of memristors that have flipped should be approximated by the probability mass below the current through the network.  This gives us a self-consistency equation of the form,
$$ f = \int_0^{\frac{g_{ON}g_{OFF}}{g_{ON} - f(g_{ON} - g_{OFF})}\frac{V}{N}} p(t) dt$$
Considering the local field $v = \frac{V}{N}$ as one of our variables, we can examine the roll of the ON-OFF ratio and distribution on the model.

```python
def current_1d(f, v):
    return g_ON * g_OFF / (g_ON - f * (g_ON - g_OFF)) * v

def self_consistency_1d_series(v, f, p):
    return quad(p, 0, current_1d(f, v))[0]
vec_self_consistency_1d_series = np.vectorize(self_consistency_1d_series)
```

So this is an increasing function of the fraction of flipped memristors and thus so is the RHS.  If we assume a uniform distribution, the self consistency equation takes the form,
$$ f =
\begin{cases}
\frac{1}{T}\frac{g_{ON}g_{OFF}}{g_{ON} - f(g_{ON} - g_{OFF})}\frac{V}{N} & \mbox{if } \frac{g_{ON}g_{OFF}}{g_{ON} - f(g_{ON} - g_{OFF})}\frac{V}{N} < T \\
1 & \mbox{if } \frac{g_{ON}g_{OFF}}{g_{ON} - f(g_{ON} - g_{OFF})}\frac{V}{N} \ge T\\
\end{cases}
$$
Resolving this for $f$ gives, 
$$f = \frac{1 - \sqrt{1 - \frac{4(1-\alpha)g_{OFF}V}{NT}}}{2(1-\alpha)}$$
for $\alpha =\frac{g_{OFF}}{g_{ON}}$. From this, the bifurcation voltage is $V_b = \frac{NT}{4(1-\alpha)g_{OFF}}$.  Similarly, the bifurcation fraction is $f_b = \frac{1}{2(1-\alpha)}$ and we immediately see we have no bifurcation for $\alpha \ge \frac{1}{2}$.

We can generalize this to a set of simultaneous equations.  The self consistency equation,
$$ f = \int_0^{G(f)V} p(t) dt$$
and an equation requiring a point of tangency,
$$ 1 = p(G(f)V)G'(f)V.$$
Using these to eliminate $V$, we find an equation for the bifurcation points.  Depending on the form of the distribution, these may or may not be analytically soluble, but probably not.

### Memristor Model

The dynamics of a memristor network depend on a number of parameters

* $R_0$ the resistance scale.  In this case we will take $R_{ON} = 0.05 R_0$ and $R_{OFF} = 1.95 R_0$
* $V_t$ the threshold voltage of our memristor(s).
* $\beta$ the proportionality constant between the derivative of the resistance and the voltage drop over the resistance for $\Delta V > V_t$
* $\alpha$ the same quantity for $V < V_t$

These give rise to the timescales

* $t_\beta = \frac{R_0}{\beta V_t}$
* $t_\alpha = \frac{R_0}{\alpha V_t}$

```python
R_0 = 1.
V_t = 1.
beta = 1.
alpha = 1e-2

R_ON = 0.05 * R_0
R_OFF = 1.95 * R_0

def XtoG(X):
    return X + X.T

def xdot(t, x, V, V_T):
    return vc_mG_xdot_dmnlss(x, V / V_T, t)

def vc_mR_xdot_dmnlss(x, V, t):
    return V - 0.5 * (np.abs(V + 1) - np.abs(V - 1))
    
def vc_mG_xdot_dmnlss(x, V, t):
    """
    If I instead want to use the conductance as the internal state variable, we have
    G_dot = -G^2 R_dot
    """
    return -x**2 * vc_mR_xdot_dmnlss(1. / x, V, t)

def vc_mG_xdot(t, x, V, V_T):
    return beta * V - 0.5 * (beta - alpha) (np.abs(V + V_T) - np.abs(V - V_T))
```

## Dimension

Perhaps by beginning with a 1D chain and gradually progressing through more and more complicated lattices, I can get a better understanding of how switching events occur

### 1D Chain

Note that 2d_cubic's method for assigning bond directions does not work for a 1d network

```python
lattice_shape = (15, 1)
num_nodes = lattice_shape[0] * lattice_shape[1]
```

```python
# Create the external voltage
t_final = 1500                                              # in units of t_0
t = np.linspace(0, t_final, 15000 )
dt = t[1] - t[0]

V_initial = np.floor(0.2 * lattice_shape[0] * V_t)
V_final = np.ceil(0.2 * R_OFF * lattice_shape[0]  / R_ON * V_t)                                    # in units of V_t
V_high = np.linspace(V_initial, V_final, len(t))
v_external = np.empty((num_nodes, len(t))) * np.nan
v_external[0:lattice_shape[1],  :] = V_high
v_external[(-1 * lattice_shape[1]):, :] = 0
```

```python
state_var = gen_adj.chain_1d(lattice_shape[0], directed=True)
threshold_voltages = state_var.copy()
rows, cols = state_var.nonzero()
R_initial = (R_OFF - R_ON) * np.random.random_sample(len(rows)) + R_ON
state_var[rows, cols] = 1. / R_initial
mnet_anis= memnets.MemristorLattice2DCubic(lattice_shape, state_var.copy(), v_external[:,0], XtoG, xdot, jacobian=None,
                           threshold_voltages=threshold_voltages, G_ON=1./R_ON, G_OFF=1./R_OFF)

mnet_anis.solve_voltages('cg')
cond = [mnet_anis.conductivity()]
# At each time point beginning with the 2nd
for t_idx, time in enumerate(t[1:]):
    # update external voltages
    mnet_anis.external_voltages = v_external[:,t_idx]
    # integrate the memristances
    mnet_anis.integrate_state_variables(dt, method='myrk4')
    # solve the network
    mnet_anis.solve_voltages('cg')
    # add the memristances to our list
    cond.append(mnet_anis.conductivity())

fig, ax = plt.subplots(figsize=(15, 12))
ax.plot(V_high, 1. / np.array(cond))
```

## Anisotropy

In a 1D chain, introducing anisotropy gives a combination of short switching events and long stabilizing tails.  In 2D we must make a choice about the anisotropy in each direction.  For starters, we will look at what happens when I merely insert no direction bias in the network.  This will be carried out for constant and random initial resistances.

```python
lattice_shape = (10, 10)
num_nodes = lattice_shape[0] * lattice_shape[1]
```

```python
# Create the external voltage
t_final = 1500                                              # in units of t_0
t = np.linspace(0, t_final, 15000 )
dt = t[1] - t[0]

V_initial = 0.5 * lattice_shape[0] * V_t
V_final = R_OFF * lattice_shape[0] * V_t                                    # in units of V_t
V_high = np.linspace(V_initial, V_final, len(t))
v_external = np.empty((num_nodes, len(t))) * np.nan
v_external[0:lattice_shape[1],  :] = V_high
v_external[(-1 * lattice_shape[1]):, :] = 0
```

### TB=1 LR=1 network

First a network in which all nodes point from left to right and top to bottom.

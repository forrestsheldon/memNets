```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
import generate_adjacency_matrix as gen_adj
import resistor_networks as rnets
from __future__ import division
```

The [literature](http://journals.aps.org/prb/pdf/10.1103/PhysRevB.34.4656) reports that a random resistor network near the percolation threshold has log-normally distributed voltage drops and a number of interesting scaling properties.  This notebook will be aimed at confirming and reviewing these findings and understanding their consequences for memristor networks.

```python
L = 512
p = 0.5
random_lattice = gen_adj.cubic_2d_random((L,L), p)
v_external = np.empty(L*L) * np.nan
v_external[0:L] = 1
v_external[-L:] = 0

rnet = rnets.ResistorNetworkCC(random_lattice, v_external)
rnet.solve_voltages('spsolve')

#fig, ax = plt.subplots(figsize=(10,10))
#rnet.display_grid(ax, (L,L), 'voltage')
rnet.conductivity()
```

```python
def calculate_voltage_drops(G, voltages):
    rows, cols = G.nonzero()
    drops_list = []
    for i, j in zip(rows, cols):
        if i < j:
            drop = np.abs(voltages[i] - voltages[j])
            if drop > 1e-12:
                drops_list.append(drop)
    return drops_list
```

```python
voltage_drops = calculate_voltage_drops(rnet.G, rnet.voltages)
len(voltage_drops)
```

```python

hist, bins = np.histogram(np.log10(voltage_drops), 150)

fig, ax = plt.subplots(figsize=(12, 12))
ax.scatter(10**bins[1:], hist)
#ax.set_yscale('log')
ax.set_xscale('log')
#ax.set_ylim(1, hist.max()*2)
ax.set_xlim((10**bins[1:].min())/2, 2*(10**bins[1:].max()))
```

```python
L = 1.
switching_times = L / np.array(voltage_drops)
switching_times.sort()
residence_times = np.ediff1d(switching_times)
residence_times = residence_times[residence_times > 1e-10]
residence_times.sort()
hist_switch, bins_switch = np.histogram(np.log10(switching_times), 300)
hist, bins = np.histogram(np.log10(residence_times[:-75]), 300)

pdf = hist / np.sum(hist)
cdf = np.cumsum(pdf)
ccdf = 1 - cdf

pdf_switch = hist_switch / np.sum(hist_switch)

fig, ax = plt.subplots(figsize = (15,12))
ax.scatter(10**bins[1:], ccdf)
ax.scatter(10**bins[1:], pdf, c='r')
ax.scatter(10**bins[1:], pdf_switch, c='g')
ax.set_xscale('log')
ax.set_yscale('log')
ax.set_xlim((10**bins[1])/1.5 , switching_times.max())
ax.set_ylim(1e-6, np.max(ccdf)*1.5)
```

```python
L = 512

v_external = np.empty(L*L) * np.nan
v_external[0:L] = 1
v_external[-L:] = 0

concentrations = [0.5, 0.55, 0.6, 0.7, 0.9, 0.99]
voltage_drops_by_concentration = []
for p in concentrations:
    #assure that the generated lattice percolates
    conductivity = 0
    while conductivity == 0:
        random_lattice = gen_adj.cubic_2d_random((L, L), p)
        rnet = rnets.ResistorNetworkCC(random_lattice, v_external)
        rnet.solve_voltages('spsolve')
        conductivity = rnet.conductivity()
    voltage_drops_by_concentration.append(calculate_voltage_drops(rnet.G, rnet.voltages))
```

```python
fig, ax = plt.subplots(figsize=(12, 12))
colors = ['r', 'g', 'b', 'y', 'c', 'm']
bin_min = 1e30
bin_max = 0
for p, voltage_drops, c in zip(concentrations, voltage_drops_by_concentration, colors):
    
    hist, bins = np.histogram(np.log10(voltage_drops), 150)
    pdf = hist / max(hist)

    ax.scatter(10**bins[1:], pdf, color=c)
    if bins[1:].min() < bin_min:
        bin_min = bins[1:].min()
    if bins[1:].max() > bin_max:
        bin_max = bins[1:].max()
    

#ax.set_yscale('log')
ax.set_xscale('log')
#ax.set_ylim(1, hist.max()*2)
ax.set_xlim((10**bin_min)/2, 2*(10**bin_max))
```
